```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris, hue='species')
```

```python
sns.pairplot(data=iris, hue='species', vars=['petal_length','petal_width'])
```

```python
sns.pairplot(data=iris, hue='species', vars=['petal_length','petal_width'], height = 4 )
```

```python
sns.jointplot(data=iris, x='petal_length', y='petal_width', kind='scatter')
```

```python
sns.jointplot(data=iris, x='petal_length', y='petal_width', kind='reg')
```

```python
sns.jointplot(data=iris, x='petal_length',y='petal_width', kind='kde')
```

```python
sns.jointplot(data=iris, x='petal_length',y='petal_width', kind='hex')
```

```python
sns.lmplot(data=iris, x='sepal_length',y='sepal_width', hue='species')
```

```python
sns.lmplot(data=iris, x='petal_length',y='petal_width', hue='species')
```
